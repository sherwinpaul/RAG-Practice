```python
from sentence_transformers import CrossEncoder

Encoder = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")

def rerank_documents(query,document,top_k=3):

    pairs= [(query,docs) for docs in document]
    scores= Encoder.predict(pairs)
    score_list=list(zip(document,scores))

    ranked_docs= sorted(score_list,key=lambda x:x[1],reverse=True)

    return ranked_docs[:top_k]
```

```python
query = "How do i reset my password?"

documents=["How to delete the account",
           "How to change my password",
           "How to create an account"]

reranked = rerank_documents(query,documents,top_k=3)

print("Reranked Documents:")

for docs,score in reranked:
    print("documents :",docs)
    print("Score :", score)
```

```python
import numpy as np

def cosine_simi(a,b):

    dotp=np.dot(a,b)

    norm_a=np.linalg.norm(a)
    norm_b=np.linalg.norm(b)

    product= dotp/(norm_a*norm_b)

    return product
```
